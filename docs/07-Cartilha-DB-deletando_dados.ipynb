## ***1.1. Deletando informações da tabela***

Iremos remover toda uma linha:

```python
def deletar_dados(con, cur, id):
    data = (id, )
    cur.execute('DELETE FROM clientes WHERE id = ?;', data)
    con.commit()
```

## ***1.2. Executando atualizações***

Como fizemos nosso código dentro de uma função, deveremos fazer da seguinte maneira:

```python
deletar_dados(con, cur, 2)
```
