# ***1. Acessando dados por meio do "Acesso direto"***

Sendo assim o modo "Acesso Direto" será permitido e, com isso, as pesquisas poderão ser feitas com um controle muito maior.

# ***2. Exemplos de apresentações possíveis com o uso do Row:*** 

Esse processo deixa o nosso trabalho mais legível. Podemos fazer vários tipos de apresentações, por exemplos.

```python
print(f"Seja bem-vindo(a) ao sistema, {cliente[1]}") # retornará o nome do usuário do índice 'id' que for 1
```

## ***2.1. Resumo sobre a importância do Row***

Um simples exemplo, mas que, ao analizarmos, abrimos um grande leque de possíbilidades para o nosso projeto ser feito de acordo com as informações do bando de dados.

O Row permite lhe garantir que a ordem em que os Atributos estão quando foram criados sejam seguidas exatamente como esta.

cliente[1] sempre será, no nosso caso, o campo 'nome', portanto retornará o resultado que foi armazenado nele nessa possição.
