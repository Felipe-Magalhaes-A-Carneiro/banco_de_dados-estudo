# ***2. Introdução aos Banco de Dados Relacionais*** 

Um BD relacional é um tipo de banco de dados que organiza os dados em tabelas (matrizes bidimensionais). Cada tabela é composta de linhas, que representam registros individuais, e colunas, que representam campos de dados e são relacionados entre si por chaves.

- Exemplos: PostgreSQL, MySQL, Oracle, SQL Server, SQLite
- Uso ideal: sistemas financeiros, ERPs, e-commerces

### ***2.1. SQL — A Linguagem dos Bancos Relacionais***

```python
-- DDL: Criar tabela
CREATE TABLE clientes (
  id       SERIAL PRIMARY KEY,
  nome     VARCHAR(100) NOT NULL,
  email    VARCHAR(150) UNIQUE,
  criado_em TIMESTAMP DEFAULT NOW()
);

-- DML: Inserir dados
INSERT INTO clientes (nome, email) VALUES ('Ana Silva', 'ana@email.com');

-- DQL: Consultar dados
SELECT c.nome, p.total
FROM clientes c
JOIN pedidos p ON p.cliente_id = c.id
WHERE p.total > 100
ORDER BY p.total DESC;

-- DDL: Alterar tabela
ALTER TABLE clientes ADD COLUMN telefone VARCHAR(20);

-- DML: Atualizar e deletar
UPDATE clientes SET telefone = '11999999999' WHERE id = 1;
DELETE FROM clientes WHERE id = 5;
```

## ***2.1. Fundamentos Teóricos:*** 

Criado por Edgar F. Codd em 1970, o modelo relacional é baseado na teoria matemática de conjuntos e álgebra relacional. Os dados são representados em relações (tabelas) compostas por:

- ***Tuplas*** → linhas (registros)
- ***Atributos*** → colunas (campos)
- ***Domínio*** → conjunto de valores válidos para um atributo

### ***2.1.1. Álgebra Relacional:***

| Operação       | Símbolo                  | Equivalente SQL                           |
| ----------- | -------------------------- | ---------------------------------- |
| Seleção    | σ | WHERE        |
| Projeção  | π     | SELECT (colunas)                       |
| União      | ∪                  | UNION                        |
| Diferença    | −  | EXCEPT |
| Produto Cartesiano | ×      | CROSS JOIN            |
| Junção | ⋈       | JOIN            |

## ***2.2. Tabelas:*** 

Em um banco dde dados relacional, uma tabela é uma estrutura organizada de linha e coluna, semelhante a uma planilha, como no Excel (mas ele não é um BD).

Cada linha representa um registro distinto e cada coluna representa um tipo de informação, chamado de campo. Por exemplo, uma tabela 'Clientes' pode ter campos como 'ID', 'nome', 'email' e 'telefone'.

# ***3. Tipos de Chaves*** 

## ***3.1. Chave Primária (PK - Primary Key):*** 

***Identifica unicamente cada registro***.

Cada tabela em um banco de dados relacional deve ter uma chave primária. A chave primátia é uma coluna (ou conjunto de colunas) cujo valor é único para cada registro.

Isso garante que cada registro na tabela possa ser identificado de maneira única.

Por exemplo, na tabela 'Clientes', o campo 'ID' pode ser a chave primária.

Isso garante a ***'unicidade'*** (caracteristica única que cada uma tem, o ID 22 é esse e não posso criar outro com esse mesmo valor) de um valor

## ***3.2. Chaves Estrangeiras (FK - Foreign Keys):*** 

***Cria relacionamento entre tabelas***.

Por exemplo, em uma tabela 'Pedidos', podemos ter um campo 'ClienteID' que seja uma chave estrangeira apontando para uma chave primária da tabela 'Clientes'.

Isso cria um relacionamento entre 'Pedidos' e 'Clientes', permitindo que cada pedido seja associado a um cliente específico.

## ***3.3. Chave Composta:*** 

***PK formada por múltiplas colunas***.

## ***4. Relacionamento entre tabelas*** 

Os bancos de dados relacionasi permitem estabelecer relações entre tabelas. As relações podem ser:

- 1:1 ('um para um') → Um usuário tem um CPF
- 1:N ('um para muitos') → Um cliente tem vários pedidos
- N:M ('muitos para muitos') → Vários alunos em vários cursos

Estas relações permitem efetuar consultas complexas que unem dados de várias tabelas.

### ***4.1. Tipos de Relacionamentos***

```python
-- 1:1 → Usuário tem um Perfil
CREATE TABLE usuarios (
  id    SERIAL PRIMARY KEY,
  login VARCHAR(50) UNIQUE NOT NULL
);

CREATE TABLE perfis (
  id         SERIAL PRIMARY KEY,
  usuario_id INT UNIQUE REFERENCES usuarios(id),
  bio        TEXT,
  avatar_url VARCHAR(255)
);

-- 1:N → Cliente tem vários Pedidos
CREATE TABLE pedidos (
  id         SERIAL PRIMARY KEY,
  cliente_id INT NOT NULL REFERENCES clientes(id),
  data       TIMESTAMP DEFAULT NOW(),
  status     VARCHAR(20) DEFAULT 'pendente'
);

-- N:M → Pedido tem vários Produtos (tabela intermediária)
CREATE TABLE itens_pedido (
  pedido_id       INT REFERENCES pedidos(id),
  produto_id      INT REFERENCES produtos(id),
  quantidade      INT NOT NULL CHECK (quantidade > 0),
  preco_unitario  NUMERIC(10,2) NOT NULL,
  PRIMARY KEY (pedido_id, produto_id)
);
```
