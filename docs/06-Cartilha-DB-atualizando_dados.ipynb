# ***Atualizando registros - SQLite***

A operação UPDATE do SQL é usada para modificar registros existentes. É importante ser específico ao usar o UPDATE para evitar alterar mais registros do que o planejado.

Portando, utilizaremos o WHERE.

Muito raramente você precisará atualizar uma tabela inteira, por tanto, com o WHERE poderemos alterar apenas o que é do nosso interesse.

Tutorial: https://docs.python.org/3/library/sqlite3.html

----------------------------------------------------------------------------------------------------

# ***1. Como fazer?*** 

```python
cur.execute()
```

Dentro de seus parâmetros, usaremos o SQL.

## ***1.1. Atualizando informações da tabela***

Dessa vez, criaremos uma função para executar o código:

```python
def atualizar_dados(con, cur, nome, email, id):
    data = (nome, email, id)
    cur.execute('UPDATE clientes SET nome= ?, email= ? WHERE id = ?;', data)
    con.commit()
```

Analizemos algumas informações:
- SET -> é de 'configurar';
- WHERE -> é de 'onde'

## ***1.2. Executando atualizações***

Como fizemos nosso código dentro de uma função, deveremos fazer da seguinte maneira:

```python
atualizar_dados(con, cur, "Felipe Magalhães", "felipe.araujo@mail.com", 1)
```

Analizemos algumas informações:
- con e cur -> Não sofreram alterações, eles são necessários;
- 'nome' atribui o meu nome e sobrenome
- 'email' será o mesmo;
- 'id' -> É aqui aqui que esta o detalhe, por eu ter dito que é '1', ele irá alterar as informações que este id possui. Dessa forma, faremos a atualização de dados mais segura. Imagine uma situação que você tenha 100 mil ids, por exemplo.
