# ***3. Utilizando SQL com Python*** 

## ***3.1. Criando(CREATE) uma nova tabela com atributos***

```python
cur.execute("CREATE TABLE clientes(id INTEGER PRIMARY KEY AUTOINCREMENT, nome VARCHAR(200), email VARCHAR(150))")
```

Obs.: Dentro do parâmetro, estamos utilizando palavras em inglês e em caixa alta, o motivo é que são exatamente essas palavras reservadas que são o SQL.

### ***3.1.1. Explicação***

- CREATE TABLE -> Cria uma nova tabela e em seguida virá o nome dessa tabela
- *nome da tabela*() -> Dentro do parênteses temos:

    Atributos:
    - id -> Será um dado do tipo INTEGER (int em Python) e será PRIMARY KEY com o uso do AUTOINCREMENT que será feito pelo próprio SQLite.
    - nome -> Será um dado do tipo VARCHAR (como String em Python), e dentro de seus parênteses o esta definido o número de caracteres que ele aceitará, no caso 200.
    - email -> Também será um dado do tipo VARCHAR, mas que permitirá apenas 150 caracteres.
