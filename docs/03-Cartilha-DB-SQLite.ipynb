# ***1. Como fazer?*** 

Em Python:

## ***1.1. Código usado***

```python
import sqlite3

con = sqlite3.connect('meu_banco_de_dados.sqlite') # a string recebe o nome do bando de dados que você quer
```

Obs.: Utilizamos a extensão 'sqlite' por conta da extensão 'SQLite Viewr' que utilizamos e ele nos obriga a fazer assim, mas você pode utilizar também a extensão 'db'.

## ***1.3. print()***

Caso utilize a função print() com a variavel escolhida para fazer a execução, por exemplo:

```python
print(con)
```

Ao executar o arquivo que possui esses códigos, perceberá que será adicinado na pasta do seu projeto um novo arquivo, com o exato nome que você deu ao banco de daddos, no meu caso, 'meu_banco_de_dados.db' e com a exata extensão que você escolheu, no meu caso, '.sqlite'.

# ***2. Aplicando o arquivo do Banco de Dados para o caminho correto*** 

## ***2.1. Uma pequena alteração***

Como é dito, ele cria automaticamente o arquivo, mas na pasta principal, porém pode ser que você queira que ele estivesse em uma pasta especifica e que sempre ocorra a conexão nela.

Basta indicar corretamente o caminho no momento que for criar a variavel de conexao, com o auxilio do 'pathlib', criando um ROOT_PATH:

```python
import sqlite3
from pathlib import Path

ROOT_PATH = Path(__file__).parent

con = sqlite3.connect(ROOT_PATH / 'meu_banco_de_dados.sqlite')
```

## ***2.2. print()***

Utilize novamente a função print() com a variavel escolhida para fazer a execução:

```python
print(con)
```

Agora o arquivo estará na pasta que também esta o arquivo que executa esse código.
