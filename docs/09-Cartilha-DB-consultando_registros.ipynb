## ***1.1. Consultando APENAS um registro***

### ***1.1.1. 'fetchone()***

O método 'fetchone()' pode ser usado para recuperar um único registro de resultado. Ele retorna o próximo registro na lista de resultados ou 'None' se nçao houver mais resultados.

```python
def consultar_cliente(cur, id):
    cur.execute('SELECT * FROM clientes WHERE id= ?', (id, ))

    return cur.fetchone()
```

- Utilizamos WHERE pois apenas um registro possuí a PK que solicitamos por meio do 'id';

- Utilizamos '*', ele representa que queremos todos os Atributos (Colunas) daquele registro com o 'id' que escolhermos, se mudarmos para 'email', por exemplo, retornará o email daquele registro 'id' que pedirmos e assim por diante.

- Perceba que não é necessário o 'con.commit()'

```python
cliente = consultar_cliente(cur, 1)

print(cliente)
```

## ***1.2. Consultando TODOS os registros***

### ***1.2.1. 'fetchall()***

O método 'fetall()' pode ser usado para recuperar todos os registros de resultados de uma vez. Ele retorna uma lista de registros ou uma lista vazia, caso não houver mais resultados.

```python
# criamos a função para retornar TODOS os clientes:
def consultar_todos_clientes(cur):
    cur.execute('SELECT * FROM clientes')
    results = cur.fetchall()

    for row in results:
        print(row)
```

```python
clientes = consultar_todos_clientes(cur)
print(clientes)
```

## ***2.3. Ordem decrescente - DESC*** 

Por padrão, o SQL retorna por ordem crescente, porém é possível alterar isso com o uso do 'DESC'.

Apenas adicione o 'DESC' depois de definir quem irá ordernar o resultado com o 'ORDER BY':

```python
def listar_por_ordem_todos_clientes(cur):
    cur.execute('SELECT * FROM clientes ORDER BY nome DESC')
    results = cur.fetchall()

    for row in results:
        print(row)
```
