# ***Inserindo registros em lote - SQLite***

'Operações em lote' como também é conhecido, são úteis quando precisamos inserir muitos registros de uma vez. Com Python DB API, podemos usar o método 'executemany()' para isso.

Tutorial: https://docs.python.org/3/library/sqlite3.html

----------------------------------------------------------------------------------------------------

# ***1. Como fazer?*** 

```python
cur.executemany()
```

Dentro de seus parâmetros, usaremos o SQL.

## ***1.1. Inserindo vários dados na tabela***

```python
def inserir_muitos_dados(con, cur, dados):
    cur.executemany('INSERT INTO clientes (nome, email) VALUES (?,?)', dados)
    con.commit()
```

## ***1.2. Executando atualizações***

Como fizemos nosso código dentro de uma função, deveremos fazer da seguinte maneira:

```python
# fazemos uma lista de tuplas:
dados = [
    ("Maria", "maria.belem@mail.com"),
    ("José", "jose.belem@mail.com"),
    ("Guilherme", "guilherme.bd@mail.com"),
    ("Ana", "ana.silvana@mail.com"),
]

inserir_muitos_dados(con, cur, dados)
```
