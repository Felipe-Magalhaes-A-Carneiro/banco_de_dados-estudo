## ***1.1. Método correto - "Querrys preparadas"***

A seguir, iremos fazer da forma correta a inserção de dados em nosso banco de dados. Esse processo é chamado de "Querrys preparadas":

### ***1.1.1. Método correto - Modelo 1"***

```python
# cria a tupla antes
data = ("Felipe", "felipe.araujo@mail.com")

cur.execute('INSERT INTO clientes (nome, email) VALUES (?, ?);', (data))
```

Perceba que depois de VALUES, definimos dentro de seus parênteses dois sinais de interrogação (já que eles representam o 'nome' e o outro o 'email').

Essa forma é a mais segura pois exige que você crie uma tupla (tipo de estrutura de dado imutável)

### ***1.1.1. Método correto - Modelo 2"***

```python
# cria os dados antes e então armazena em uma tupla
nome = "Felipe"
email = "felipe.araujo@mail.com"

cur.execute('INSERT INTO clientes (nome, email) VALUES (?, ?);', (nome, email))
```

# ***2. Método errado:***

Talvez você encontre alguém que utilize o concatenar dados. Essa não é a forma correta de fazer o código:

```python
# cria os dados antes
nome = "Felipe"
email = "felipe.araujo@mail.com"

# Depois concatena com 'f' de format:
cur.execute(f'INSERT INTO clientes (nome, email) VALUES ({nome}, {email})')
```

Essa forma esta incorreta porque ela é insegura.

Alguém pode ir ao seu código, e em alguma das váriaveis criar uma condição como TRUE e então criar uma resposta ao Python aceita-lo, por exemplo:

```python
nome = "Felipe"
email = "felipe.araujo@mail.com 1=1 DROP TABLE" # o SQL aceita, diz que 1 é igual a 1, logo TRUE e, se é TRUE: Derrube a tabela.

cur.execute(f'INSERT INTO clientes (nome, email) VALUES ({nome}, {email})')
```

***Lembre-se: Essa forma é errada de se programar a inserção de dados em uma tabela, podendo causar dados graves ao projeto!***
